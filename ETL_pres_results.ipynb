```python
from pathlib import Path
import pandas as pd
```

```python
#read in CSV of presidential restuls from 2000 through 2020
data = Path('/Users/kristin/Documents/Project_4/Resources/countypres_2000-2020.csv')
df = pd.read_csv(data)
df.head()
```

```python
#drop unnecessary columns
df_dropped = df.drop(['state_po', 'county_fips', 'version', 'mode'], axis=1)
df_dropped
```

```python
#drop parties other than democrat or republican
df_binary = df_dropped[df_dropped['party'].isin(['DEMOCRAT', 'REPUBLICAN'])]
df_binary
```

```python
#create df for just PA results
pa_df = df_binary.loc[(df_binary['state'] == "PENNSYLVANIA")]
pa_df
```

```python
# group by year, state, and county_name and find the candidate with the highest votes
pa_winners_df = pa_df.loc[pa_df.groupby(['year', 'state', 'county_name'])['candidatevotes'].idxmax()]

# Display the winners
pa_winners_df
```

```python
#drop state column
pa_winners_df = pa_winners_df.drop(['state', 'office', 'candidate', 'candidatevotes', 'totalvotes'], axis=1)
pa_winners_df
```

```python
#change the party column data to 0 and 1 
pa_winners_df['party'] = pa_winners_df['party'].replace({'DEMOCRAT': 0, 'REPUBLICAN': 1})
pa_winners_df
```

```python
#rename column name for winning party for clarification 
pa_winners_df = pa_winners_df.rename(columns = {'party':'winning_party D:0 R:1'})
pa_winners_df
```

```python
#separate out winners for 2012
pa_winners_df_2012 = pa_winners_df[pa_winners_df['year'] == 2012]
pa_winners_df_2012
```

```python
# Drop rows for counties that we do not have iformation on 
rows_to_drop = ['BEDFORD', 'BRADFORD', 'CAMERON', 'CLARION', 'CLINTON', 'ELK', 'FOREST',
                'FULTON', 'GREENE', 'HUNTINGDON', 'JEFFERSON', 'JUNIATA', 'MCKEAN', 'MIFFLIN', 'MONTOUR',
                'PERRY', 'PIKE', 'POTTER', 'SNYDER', 'SULLIVAN', 'SUSQUEHANNA', 'TIOGA',
                'UNION', 'VENANGO', 'WARREN', 'WAYNE', 'WYOMING']

#print the cleaned dataframe and reset the index
cleaned_pa_winners_df_2012 = pa_winners_df_2012[~pa_winners_df_2012['county_name'].isin(rows_to_drop)]
cleaned_pa_winners_df_2012 = cleaned_pa_winners_df_2012.reset_index(drop=True)
cleaned_pa_winners_df_2012
```

```python
#separate out winners for 2016
pa_winners_df_2016 = pa_winners_df[pa_winners_df['year'] == 2016]
pa_winners_df_2016
```

```python
# Drop rows for counties that we do not have iformation on 
rows_to_drop = ['BEDFORD', 'BRADFORD', 'CAMERON', 'CLARION', 'CLINTON', 'ELK', 'FOREST',
                'FULTON', 'GREENE', 'HUNTINGDON', 'JEFFERSON', 'JUNIATA', 'MCKEAN', 'MIFFLIN', 'MONTOUR',
                'PERRY', 'PIKE', 'POTTER', 'SNYDER', 'SULLIVAN', 'SUSQUEHANNA', 'TIOGA',
                'UNION', 'VENANGO', 'WARREN', 'WAYNE', 'WYOMING']

#print the cleaned dataframe and reset the index
cleaned_pa_winners_df_2016 = pa_winners_df_2016[~pa_winners_df_2016['county_name'].isin(rows_to_drop)]
cleaned_pa_winners_df_2016 = cleaned_pa_winners_df_2016.reset_index(drop=True)
cleaned_pa_winners_df_2016
```

```python
#separate out winners for 2020
pa_winners_df_2020 = pa_winners_df[pa_winners_df['year'] == 2020]
pa_winners_df_2020
```

```python
# Drop rows for counties that we do not have iformation on 
rows_to_drop = ['BEDFORD', 'BRADFORD', 'CAMERON', 'CLARION', 'CLINTON', 'ELK', 'FOREST',
                'FULTON', 'GREENE', 'HUNTINGDON', 'JEFFERSON', 'JUNIATA', 'MCKEAN', 'MIFFLIN', 'MONTOUR',
                'PERRY', 'PIKE', 'POTTER', 'SNYDER', 'SULLIVAN', 'SUSQUEHANNA', 'TIOGA',
                'UNION', 'VENANGO', 'WARREN', 'WAYNE', 'WYOMING']

#print the cleaned dataframe and reset the index
cleaned_pa_winners_df_2020 = pa_winners_df_2020[~pa_winners_df_2020['county_name'].isin(rows_to_drop)]
cleaned_pa_winners_df_2020 = cleaned_pa_winners_df_2020.reset_index(drop=True)
cleaned_pa_winners_df_2020
```

```python
#merge the three years together
combined_county_results = pd.concat([cleaned_pa_winners_df_2012, cleaned_pa_winners_df_2016, cleaned_pa_winners_df_2020], ignore_index=True)
combined_county_results
```

```python
#clean up the column names to match the county attributes df that it will be joined with
combined_county_results_clean = combined_county_results.rename(columns = {'year':'Year', 'county_name': 'County'})
combined_county_results_clean
```

```python
# Export the DataFrame as a CSV file. 
combined_county_results_clean.to_csv("Resources/combined_county_results_clean.csv", encoding='utf8', index=False)
```
