This notebook contains code regarding the final merging of all complete county attribute datasets (2012, 2016, 2019 & 2023) and subsequent export for futher visualization.

```python
import pandas as pd
```

```python
# Read in the combined 2012, 2016, 2019 dataset and the 2023 dataset for merge
# 2023 dataset 
county_2023 = pd.read_csv("Resources/2023_County_Final_With_Segments_And_Predictions.csv")
county_2023
```

```python
# Swap position of 'Segment' and 'winning_party D:0 R:1' for concatenation alignment
# Switch columns 103 and 104
# Get the current column names
columns = county_2023.columns.tolist()

# Switch the positions of the 103rd and 104th columns
columns[102], columns[103] = columns[103], columns[102]

# Reorder the DataFrame & display
county_2023 = county_2023[columns]
county_2023
```

```python
# Combined 2012, 2016, 2019 dataset
county_2012_2016_2019 = pd.read_csv("Resources/County_Final_With_Outcomes_And_Segments.csv")
county_2012_2016_2019
```

```python
# Concatenate both datasets vertically to form a single complete county dataset
combined_df = pd.concat([county_2012_2016_2019, county_2023], ignore_index=True)
combined_df
```

```python
# Export
combined_df.to_csv("Resources/Complete_County_Dataset.csv")
```
